```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
colnames = ['population', 'profit']
df= pd.read_csv("ex1data1.txt",names =colnames)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.plot(x="population", y="profit", style = 'o')
plt.show()
```

```python
#population = [population]
#profit = [profit]

# Fit a linear regression line
coefficients = np.polyfit(df.population, df.profit, 1)
line = np.polyval(coefficients, df.population)

# Plot the scatter plot
plt.scatter(df.population, df.profit, label='Data points')

# Plot the best fit line
plt.plot(df.population, line, color='red', label='Best Fit Line')

# Add labels and title
plt.xlabel('Population')
plt.ylabel('Profit')
plt.title('Scatter Plot with Best Fit Line')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
# Print the coefficients (parameters)
print("Coefficients:", coefficients)

# Calculate the cost function (mean squared error)
cost_function = np.mean((line - df.profit) ** 2)
print("Cost Function (Mean Squared Error):", cost_function)
```

Gradient descent simply is an algorithm that makes small steps along a function to find a local minimum. We can look at a simply quadratic equation such as this one

```python
# Generate some random data
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Add a bias term to X
X_b = np.c_[np.ones((100, 1)), X]

# Set the learning rate and number of iterations
learning_rate = 0.004
n_iterations = 1500

# Initialize random theta values
theta = np.random.randn(2, 1)

# Perform gradient descent
for iteration in range(n_iterations):
    gradients = 2 / 100 * X_b.T.dot(X_b.dot(theta) - y)
    theta = theta - learning_rate * gradients

# Print the final theta values
print("Final Theta:", theta)

# Plot the data points and the best fit line
plt.scatter(X, y, label='Data points')
plt.plot(X, X_b.dot(theta), color='red', label='Best Fit Line')
plt.xlabel('X')
plt.ylabel('y')
plt.title('Linear Regression with Gradient Descent')
plt.legend()
plt.show()
```

```python
# Calculate the predicted values
y_pred = X_b.dot(theta)

# Calculate the mean squared error (MSE)
mse = np.mean((y_pred - y) ** 2)

# Print the MSE
print("Mean Squared Error (MSE):", mse)
```
